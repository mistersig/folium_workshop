```python
import os
import folium
import pandas
```

```python
speedPoints = os.path.join('data', 'speed_camera_violations.csv')
scamera = pandas.read_csv(speedPoints)
scamera.head()
```

```python
scamera.count()
```

```python
scamera = scamera[:100]

locations = scamera[['LATITUDE','LONGITUDE']]
locationlist = locations.values.tolist()
locationlist
```

```python
m = folium.Map(location=[41.84510,-87.630232], tiles='CartoDB positron', zoom_start=11)
m
```

```python
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=scamera['CAMERA_ID'][point]).add_to(m)
m
```

```python
m = folium.Map(location=[41.84510,-87.630232], tiles='Stamen Toner', zoom_start=11)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], 
    popup=scamera['CAMERA_ID'][point],
    icon = folium.Icon(color='green',icon_color='#fff', icon='camera')).add_to(m)
m
```
